```python
shopping_df = pd.read_csv('../DATA/shopping_data.csv')
shopping_df.head()
```

```python
shopping_df.dtypes
```

```python
for column in shopping_df.columns:
    print(f'Column {column} has {shopping_df[column].isnull().sum()} null values')
```

```python
shopping_dfclean = shopping_df.dropna()
shopping_dfclean.head()
```

```python
shopping_dfclean.drop(columns = ['CustomerID'], inplace = True)
```

```python
shopping_dfclean.head()
```

```python
def change_string(member):
    if member == 'Yes':
        return 1
    else:
        return 0
    

shopping_dfclean['Card Member'] = shopping_dfclean['Card Member'].apply(change_string)
shopping_dfclean.head()
```

```python
shopping_dfclean['Annual Income'] = shopping_dfclean['Annual Income']/1000
shopping_dfclean.head()
```

```python
shopping_dfclean = shopping_dfclean.rename(columns={'Card Member':'Card_Member','Age':'Age','Annual Income':'Annual_Income','Spending Score (1-100)':'Spending_Score'})
shopping_dfclean.head()
```

```python
shopping_dfclean.to_csv('../DATA/shopping_clean.csv', index = False)
```
